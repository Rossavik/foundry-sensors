```python
import pandas as pd

data = pd.read_csv('processed_data.csv')

# Display the first few rows of the dataframe to understand its structure
data.head(), data.columns
```

```python
# Selecting the required columns and creating the new DataFrame with conditions
required_data = data[['regnr', 'datostøpt', 'datoVrakReg', 'antall vrak', 'vrakårsak']].copy()
required_data.columns = ['regnr', 'datostøpt', 'datoVrakReg', 'antall vrak', 'vrakårsak']

# Apply the conditions for 'godkjent' and 'cast error' columns
required_data['godkjent'] = required_data['antall vrak'] == 0
required_data['cast error'] = required_data['antall vrak'] != 0

# Convert the DataFrame to a CSV file
required_data.to_csv('cast_error_data.csv', index=False)

required_data.head()
```

```python
# Get unique values of 'vrakårsak'
unique_vrakårsak = required_data['vrakårsak'].dropna().unique()
#unique_vrakårsak

# Create an indexed list of unique 'vrakårsak' values
indexed_vrakårsak = pd.Series(unique_vrakårsak).sort_values().reset_index(drop=True).reset_index()
indexed_vrakårsak.columns = ['Index', 'vrakårsak']
indexed_vrakårsak

```
